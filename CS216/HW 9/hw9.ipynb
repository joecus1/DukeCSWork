# Part 2: Visualization Practice Problems
You may complete the following exercises using any of the visualization tools described above: Matplotlib, Pandas, or Seaborn. We will work through some visualizations of a dataset containing standard specifications about automobiles produced in the United States, Japan, and Europe between 1970 and 1982. We begin by loadig this dataset into a pandas dataframe called `cars`. 

```python
# This line is just to make matplotlib play nicely with jupyter notebooks
%matplotlib inline 

# Importing the relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

# Importing the dataset we will work with
cars = sns.load_dataset('mpg')
cars.head()
```

### Problem A
Use a single bar plot to visualize the average mpg of cars of the three different origins: `usa`, `japan`, and `europe`.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
usaMPGTot = 0
usaTot = 0
japanMPGTot = 0
japanTot = 0
europeMPGTot = 0
europeTot = 0
usaMPG = []
japanMPG = []
europeMPG = []
for i in range(len(cars["origin"])):
    if cars["origin"][i] == "usa":
        usaMPGTot += cars["mpg"][i]
        usaTot += 1
        usaMPG.append(cars["mpg"][i])
    if cars["origin"][i] == "japan":
        japanMPGTot += cars["mpg"][i]
        japanTot += 1
        japanMPG.append(cars["mpg"][i])
    if cars["origin"][i] == "europe":
        europeMPGTot += cars["mpg"][i]
        europeTot += 1
        europeMPG.append(cars["mpg"][i])
usaAVG = usaMPGTot / usaTot
japanAVG = japanMPGTot / japanTot
europeAVG = europeMPGTot / europeTot
countries = ["usa", "japan", "europe"]
mpg = [usaAVG, japanAVG, europeAVG]
ax.bar(countries,mpg)
plt.show()
```

### Problem B
The bar plot from problem A shows the averages in a simple format, but what if we would like more information about the distribution? Use a single box plot to visualize the the distribution of mpg for each of the three origins of car (`usa`, `japan`, and `europe`).

```python
sns.boxplot(x="origin", y="mpg", data = cars)
```

### Problem G
By now you should have developed some intuition for the nuances of this dataset.
